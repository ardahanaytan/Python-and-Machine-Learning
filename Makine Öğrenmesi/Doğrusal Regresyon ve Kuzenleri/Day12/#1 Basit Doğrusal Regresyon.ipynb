```python
import statsmodels.api as sm
```

```python
X = df[['TV']]
X[0:5]
```

```python
X = sm.add_constant(X)
```

```python
X[0:5]
```

```python
y = df['sales']
```

```python
y[0:5]
```

```python
lm = sm.OLS(y,X)
```

```python
model = lm.fit()
```

```python
model.summary()
```

```python
import statsmodels.formula.api as smf
lm = smf.ols('sales ~ TV', df)
model = lm.fit()
model.summary()
```

```python
model.params
```

```python
model.summary().tables[1]
```

```python
model.conf_int()
```
